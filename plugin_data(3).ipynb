```
import pandas as pd
import numpy as np
from tqdm import tqdm
path  = '/content/drive/My Drive/data/plugin_data/'
```

```
pro_plug = pd.read_csv(path+'processed_plugin.csv')
tags = pd.read_csv(path+'tags.csv')
```

```
low_mar = pro_plug[(pro_plug['install_count'] > 0) & ((pro_plug['avg_down'] > 30) & (pro_plug['avg_down'] < 100))]
```

```
tags = tags[tags['num_plugins'] > 0]
tags
```

```
def is_tag_in_low_mar(tag_id):
  for tags in low_mar['tags']:
    if str(tag_id) in tags.split(' '):
      return True
  return False

low_tags = tags[tags['id'].apply(is_tag_in_low_mar)]
```

```
low_tags['name'] = low_tags['name'].str.lower()
```

```
low_tags
```

```
tags_keywords = []
for name in low_tags['name']:
  if "wordpress" in name.lower():
    tags_keywords.append(name+" plugins")
  else:
    tags_keywords.append("wordpress "+name+" plugins")
```

```
splt_array = np.array_split(tags_keywords,4)
```

```
for elm in splt_array:
  print(len(elm))
```

```
with open("tags_list.txt","w") as f:
  for name in tags_keywords:
    f.write(name+"\n")
    
```

```
i = 0
for elm in splt_array:
  with open("tags_list_"+str(i)+".txt",'w') as f:
    for name in elm:
      f.write(name+"\n")
  i+=1
```
